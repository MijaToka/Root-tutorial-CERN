Create a histogram, fill it with random numbers, set its colour to blue, draw it.

```python
import ROOT
from random import gauss,random
import numpy as np
```

```python
nEvents = 1000
events = np.array([gauss() for _ in range(nEvents)])
events
```

```python
nbins,low,up = 30,-6,6
h = ROOT.TH1D("h","Histogram;;N Events",nbins,low,up)
for event in events:
    h.Fill(event)
```

```python
%jsroot
if ROOT.gROOT.FindObject('canvas'):
    ROOT.gROOT.FindObject('canvas').Close()

canvas = ROOT.TCanvas('canvas','Random distriburtion',400,300)

h.SetLineColor(ROOT.kBlue)
h.SetFillColor(ROOT.kBlue)


h.Draw()
canvas.Draw()
```
